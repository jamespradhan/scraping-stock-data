```python
import pandas as pd
import yfinance as yf
```

```python
d = pd.read_csv("data/index-constituent/nifty-50.csv")
d.head()
```

```python
tickers = d["symbol"].tolist()
print(tickers)
```

```python
data = yf.download(tickers, interval="1d", period="max", auto_adjust=True)
data
```

```python
file_name = "nifty-50.csv"
data.to_csv(f"data/index-stocks-data/{file_name}", index=True)
```
